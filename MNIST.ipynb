# Description

MNIST, the Modified National Institute of Standards and Technology, is a sub-set database of handwritten digits derived from the larger NIST database. It is used popularly to train machine learning models to recognize handwritten digits. 

The MNIST database has 70,000 examples, which is split into a training set and a test set. The first 60 000 examples are included in the training set, whilst the last 10 000 examples are part of the test set. Each example has an image (28x28 pixel greyscale) and a label of digits ranging from 0 to 9 to be used for training and testing of a neural network, for example, a feed-forward network or multi-layer perceptron, to classify the digits.

# Section 1: Basic Neural Network
### Import Libraries and Preprocess Data

NumPy is needed for numerical calculations, Matplotlib for plotting graphs and TensorFlow and Keras to build the neural network.

We need to convert the pixel scale values of the training and testing images from 0 - 255 to 0 - 1 after loading the data from the mnist database. We do this by dividing by 255. 

The initial format of the images is 3D (number of samples, image rows, image columns). Because feed-forward neural networks take in 1D input, we need to reshape them into vectors (28 * 28 = 784 = rows * cols).

Additionally, we need to adapt the training and testing labels into vectors to allow us to work with them in 10 categories (digits from 0 to 9).

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.metrics import confusion_matrix

import tensorflow as tf
from tensorflow import keras
from keras import datasets
from keras import utils

(x_train_images, train_labels), (x_test_images, test_labels) = datasets.mnist.load_data()

# convert pixel scale values
train_images = x_train_images / 255
test_images = x_test_images / 255

num_samples_train, img_rows, img_cols = train_images.shape
train_images = train_images.reshape(num_samples_train, img_rows * img_cols)

num_samples_test, _, _ = test_images.shape
test_images = test_images.reshape(num_samples_test, img_rows * img_cols)

lbl_train = utils.to_categorical(train_labels, 10) 
lbl_test = utils.to_categorical(test_labels, 10) 
```

## Build the Neural Network and Train the Model

### Starting Small | 100 Samples

```python
model = tf.keras.Sequential([
    keras.layers.Input(shape=(28*28,)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_images[:100], lbl_train[:100], epochs=16, batch_size=32, verbose=1)
```

### Evaluate the Model

To determine the model's performance, we will evaluate it's accuracy and make predictions for the first 20 test images, comparing the predicted digits to the true digit labels.

```python
loss, accuracy = model.evaluate(test_images, lbl_test)
base_accuracy = round(accuracy*100, 2)
print(f"Test accuracy: {base_accuracy}%")
acc_list = []
acc_list.append(base_accuracy)

predictions = model.predict(test_images[:20,:])
predictions_idx = predictions.argmax(1)
print(f"Predictions: {predictions_idx}")
print(f"Actual:      {test_labels[:20]}")
```

### Some Prediction Errors

The model makes some errors in predicting digits. However, this is exmpected with such a small sample size. Let's see how the accuracy changes when increasing the training set.

### 500 Samples

### Full Training Sample | 60000 Samples

```python
model = tf.keras.Sequential([
    keras.layers.Input(shape=(28*28,)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_images, lbl_train, epochs=16, batch_size=32, verbose=1)
```

### Confusion Matrix of Full Set

A confusion matrix allows us to analyze the accuracy of our model in predicting each digit by breaking it up into predicted versus actual labels. In this way, we are able to see specifically where our model is making mistakes, including false negatives and false positives (which negatively affect our accuracy score).

We can derive an accuracy score for each category (0 to 9) by dividing the number of true positives by the total samples of that specific category. See the confusion matrix and breakdown of accuracy below:

```python
loss, accuracy = model.evaluate(test_images, lbl_test)
accuracy = round(accuracy*100, 2)
print(f"Test accuracy: {accuracy}%")
acc_list.append(accuracy)

predictions = model.predict(test_images)
predictions_idx = predictions.argmax(1)
true_labels = lbl_test.argmax(1)

# Confusion Matrix
conf_matrix = confusion_matrix(true_labels, predictions_idx)

plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Purples)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(10)
plt.xticks(tick_marks, tick_marks)
plt.yticks(tick_marks, tick_marks)

# Add labels to each square
thresh = conf_matrix.max() / 2
for i, j in np.ndindex(conf_matrix.shape):
    plt.text(j, i, format(conf_matrix[i, j], 'd'),
             horizontalalignment="center",
             color="white" if conf_matrix[i, j] > thresh else "black")

plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.tight_layout()
plt.show()

print(f"Predictions: {predictions_idx[:20]}")
print(f"Actual:      {test_labels[:20]}")
```

```python
# calculating the accuracy of each digit
class_accuracies = {}
for i in range(10):  # 0-9 digits
    true_positives = conf_matrix[i, i]
    total_samples = np.sum(conf_matrix[i, :])
    class_accuracies[i] = (true_positives / total_samples) * 100

# Extract the numbers (0-9) and accuracies
numbers = list(class_accuracies.keys())
accuracies = list(class_accuracies.values())

plt.bar(numbers, accuracies, color='pink')
plt.title('Accuracy per Number for MNIST')
plt.xlabel('Number (Digit)')
plt.ylabel('Accuracy (%)')
plt.ylim(80, 100) 
plt.xticks(numbers)
plt.show()
```

# Section 2: Establishing the Experimental Environment

```python
# lists for each accuracy experiment used for graphing
acc_list_neurons = []
acc_list_layers = []
acc_list_epochs = []
acc_list_batches = []
acc_list_afs = []

# lists for each training time experiment used for graphing
elapsed_neurons = []
elapsed_layers = []
elapsed_epochs = []
elapsed_batches = []
elapsed_afs = []

def create_model(neuron_num, num_layer, af):
    model = tf.keras.Sequential()
    model.add(keras.layers.Input(shape=(28*28,)))

    for i in range(num_layer):
        model.add(keras.layers.Dense(neuron_num, activation=af))
            
    model.add(keras.layers.Dense(10, activation='softmax'))

    model.compile(optimizer='adam',
                loss='categorical_crossentropy',
                metrics=['accuracy'])

    return model

def train(mdl, ep_mode, epoch_num, batch_num):
    start = time.perf_counter()
    mdl.fit(train_images, lbl_train, epochs=epoch_num, batch_size=batch_num, verbose=1)
    stop = time.perf_counter()

    if ep_mode == 0: # testing neurons
        elapsed_neurons.append(round(stop - start, 2))
    elif ep_mode == 1: # testing layers
        elapsed_layers.append(round(stop - start, 2))
    elif ep_mode == 2: # testing epochs
        elapsed_epochs.append(round(stop - start, 2))
    elif ep_mode == 3: # testing batches
        elapsed_batches.append(round(stop - start, 2))
    elif ep_mode == 4: # testing activation functions
        elapsed_afs.append(round(stop - start, 2))
    
def evaluate_model(neuron_num, num_layer, epoch_num, ep_mode, batch_num, af):
    model = create_model(neuron_num, num_layer, af)
    train(model, ep_mode, epoch_num, batch_num)
    loss, accuracy = model.evaluate(test_images, lbl_test)
    accuracy = round(accuracy*100, 2)
    if ep_mode == 0:
        acc_list_neurons.append(accuracy)
    elif ep_mode == 1:
        acc_list_layers.append(accuracy)
    elif ep_mode == 2:
        acc_list_epochs.append(accuracy)
    elif ep_mode == 3:
        acc_list_batches.append(accuracy)
    elif ep_mode == 4:
        acc_list_afs.append(accuracy)

def graph_accuracy(x_ax, y_ax, ex_field):
    plt.figure(figsize=(10,6))
    plt.plot(x_ax, y_ax, marker='o')
    plt.xlabel(f'{ex_field}')
    plt.ylabel('Accuracy (%)')
    plt.title(f'Effect of {ex_field} on Accuracy')
    plt.grid(True)
    plt.show()

def graph_time(x_ax, y_ax, ex_field):
    plt.figure(figsize=(10,6))
    plt.plot(x_ax, y_ax, marker='o')
    plt.xlabel(f'{ex_field}')
    plt.ylabel('Training Time (s)')
    plt.title(f'Effect of {ex_field} on Training Time')
    plt.grid(True)
    plt.show()
```

## EXPERIMENT 1 | Neurons

<b>Hypothesis</b>: More neurons per layer will result in higher accuracy and longer execution time; however, too many neurons may lead to overfitting.

```python
neuron_nums = [10, 40, 80, 160] # testing 4 models with 10, 40, 80 and 160 neurons per hidden layer of each model

for neuron_num in neuron_nums:
    evaluate_model(neuron_num, 2, 4, 0, 8, "relu")

print(acc_list_neurons)
print(elapsed_neurons)

graph_accuracy(neuron_nums, acc_list_neurons, "Number of Neurons")
graph_time(neuron_nums, elapsed_neurons, "Number of Neurons")
```

## Experiment 1 | Observations

The number of neurons per hidden layer of the model is directly proportional to the accuracy and training time of the model. I tested multiple different scenarios for neuron numbers, but only saw a decrease in accuracy due to overfitting with 160 neurons and a batch size of 8 in one test. Otherwise, the accuracy graph displayed a regular increasing curve. Overfitting due to number of neurons is not independent of the other parameters.

## EXPERIMENT 2 | Hidden Layers

<b>Hypothesis</b>: More hidden layers will result in higher accuracy and longer execution time; however, too many layers may lead to overfitting.

```python
num_layers = [2, 5, 7, 10] # testing 4 models with 2, 5, 7, and 10 hidden layers respectively

for num_layer in num_layers:
    evaluate_model(10, num_layer, 4, 1, 32, "relu")

print(acc_list_layers)
print(elapsed_layers)

graph_accuracy(num_layers, acc_list_layers, "Number of Hidden Layers")
graph_time(num_layers, elapsed_layers, "Number of Hidden Layers")
```

## Experiment 2 | Observations

My hypothesis was incorrect. The number of hidden layers in a neural network is inversely proportional to the accuracy of the model, but directly proportional to the training time of the model.

## EXPERIMENT 3 | Epochs

<b>Hypothesis</b>: More epochs will result in higher accuracy and longer execution time; however, too many epochs may lead to overfitting.

```python
epoch_nums = [4, 8, 16, 32]

for epoch_num in epoch_nums:
    evaluate_model(10, 2, epoch_num, 2, 32, "relu")

print(acc_list_epochs)
print(elapsed_epochs)

graph_accuracy(epoch_nums, acc_list_epochs, "Number of Epochs")
graph_time(epoch_nums, elapsed_epochs, "Number of Epochs")
```

## EXPERIMENT 4 | Batch Size

<b>Hypothesis</b>: A larger batch size will result in higher accuracy and longer execution time; however, too large of a batch size may lead to overfitting.

```python
batch_nums = [16, 64, 128, 256]

for batch_num in batch_nums:
    evaluate_model(10, 2, 4, 3, batch_num, "relu")

print(acc_list_batches)
print(elapsed_batches)

graph_accuracy(batch_nums, acc_list_batches, "Batch Size")
graph_time(batch_nums, elapsed_batches, "Batch Size")
```

```python
def graph_accuracy_and_time(x_ax, y_acc, y_time, ex_field):
    fig, ax1 = plt.subplots(figsize=(10,6))

    # Plot accuracy (left y-axis)
    ax1.set_xlabel(f'{ex_field}')
    ax1.set_ylabel('Accuracy (%)', color='tab:blue')
    ax1.plot(x_ax, y_acc, marker='o', color='tab:blue', label='Accuracy')
    ax1.tick_params(axis='y', labelcolor='tab:blue')
    ax1.grid(True)

    # Create a second y-axis for training time
    ax2 = ax1.twinx()  
    ax2.set_ylabel('Training Time (s)', color='tab:red')
    ax2.plot(x_ax, y_time, marker='o', color='tab:red', label='Training Time')
    ax2.tick_params(axis='y', labelcolor='tab:red')

    # Title and layout adjustments
    fig.tight_layout()  
    plt.title(f'Effect of {ex_field} on Accuracy and Training Time')

    plt.show()

# Example usage
batch_nums = [16, 64, 128, 256]

for batch_num in batch_nums:
    evaluate_model(10, 2, 4, 3, batch_num, "relu")

print(acc_list_batches)  # accuracy for each batch size
print(elapsed_batches)   # time for each batch size

graph_accuracy_and_time(batch_nums, acc_list_batches, elapsed_batches, "Batch Size")
```

## Experiment 4 | Observations

My hypothesis was incorrect. Batch size is inversely proportional to the accuracy and training time of the neural network.

## Additional Experimentation
Implementing a 5-Fold Cross-Validation technique can significantly improve model evaluation while reducing overfitting the data. Let's attempt an example:


```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True)
accuracy_per_fold = []
loss_per_fold = []

for train_idx, val_idx in kf.split(train_images):
    model = tf.keras.Sequential([
        keras.layers.Input(shape=(28*28,)),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(10, activation='softmax')
    ])

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    
    history = model.fit(train_images[train_idx], lbl_train[train_idx], 
                        validation_data=(train_images[val_idx], lbl_train[val_idx]), 
                        epochs=16, batch_size=32, verbose=1)
    
    loss, accuracy = model.evaluate(train_images[val_idx], lbl_train[val_idx])
    accuracy_per_fold.append(accuracy)
    loss_per_fold.append(loss)

avg_accuracy = np.mean(accuracy_per_fold)
avg_loss = np.mean(loss_per_fold)
```

```python
# Accuracy plot for 5FCV
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(range(1, 6), accuracy_per_fold, color='lightblue')
plt.title('Accuracy per Fold')
plt.xlabel('Fold')
plt.ylim(0.80, 1) 
plt.ylabel('Accuracy')

plt.show()
```


# Conclusions
As the number of neurons and epochs increases, the accuracy of the neural network model increases. This is because an increase in these components improves the ability for the neural network to train. Specifically, the increase in the number of neurons increases the capacity to learn complex patterns, allows it to form more complex representations of the data and, overall, improves the learning of the model. An increase in the number of epochs allows for the model to train more, and more training gives more accurate results. It also helps to prevent underfitting (when the model does not train long enough to pick up accurate patterns).

<b>Overfitting</b>: I was expecting a visible drop in accuracy after a certain "optimal" point in the number of neurons per hidden layer and epochs, indicating that the model trained "too well" and overfitted for the test set. However, I was unable to push these numbers past the optimal point to where overfitting occurs due to hardware exhaustion (memory), so I adjusted my parameters as my resources allowed to explore the general trends. In a perfect environment, we can expect overfitting to occur with too many neurons and epochs.

As the number of hidden layers and batch size increases, the accuracy of the neural network model decreases. This is due to overfitting due to too many hidden layers in the neural network making it too complex to be able to learn the general patterns while training, and encountering noise. Very large batch sizes can worsen the model's ability to generalize during training, leading to lower accuracy.

<b>Training time</b>: As the number of neurons, hidden layers, and epochs increases, the training time of the neural network model increases, while with an increase in batch size, the training time decreases. Training time decreases for an increased batch size due to parellelization (simultaneous processing), fewer updates, and reduced overhead (data is processed in fewer, larger chunks).

<b>Activation functions</b>: ReLU seems slightly more effective than Tanh, with a slightly higher accuracy and faster training time; however this difference is not sigificant.

Additional experiementation involves implementing 5-Fold Cross-Validation. Due to hardware constraints, the full effect of this alternate test compared to the basic testing will not be observable on this scale. Nevertheless, 5FCV should prove to reduce overfitting significantly.
